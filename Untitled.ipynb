```python
data['Pitch class F#'] == 'YES'
```

```python
plt.figure(figsize = (10, 6))

plt.scatter(data[data['Pitch class F#'] == 'YES']['Event number'],
            data[data['Pitch class F#'] == 'YES']['Meter'],
            alpha = 0.05,
            label = 'YES',
            color = 'b')

plt.scatter(data[data['Pitch class F#'] == 'NO']['Event number'],
            data[data['Pitch class F#'] == 'NO']['Meter'],
            alpha = 0.75,
            label = 'NO',
            color = 'r')

plt.xlabel('Event number')
plt.xticks(np.arange(0, 207, step=50))
plt.xlim(0, 207)
plt.ylabel('Meter')
plt.legend()
#plt.grid()
```

```python
np.random.seed(0)
education_num_rnd = data['Event number'] + np.random.rand(data.shape[0]) - .5

plt.figure(figsize = (10, 8))

plt.scatter(education_num_rnd[data['Pitch class F#'] == 'YES'],
            data[data['Pitch class F#'] == 'YES']['Meter'],
            alpha = 0.15,
            label = 'YES',
            color = 'b')

plt.scatter(education_num_rnd[data['Pitch class F#'] == 'NO'],
            data[data['Pitch class F#'] == 'NO']['Meter'],
            alpha = 0.15,
            label = 'NO',
            color = 'r')

plt.xlabel('Event number')
plt.xticks(np.arange(0, 207, step=50))
plt.xlim(0, 207)
plt.ylabel('Meter')
plt.legend()
#plt.grid()
plt.title('From Census Data of 1994')
```
